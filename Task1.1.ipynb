```python
import matplotlib.pyplot as plt

# Define the columns you want to plot (assuming you want to plot columns from the 5th to the last)
columns_to_plot = df.columns[4:]

# Number of columns and rows for subplots
n_cols = 4
n_rows = (len(columns_to_plot) + n_cols - 1) // n_cols  # Ensures enough rows to fit all subplots

# Create a figure and subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 45))
fig.tight_layout(pad=3.0)

# Flatten axes array for easy iteration
axes = axes.flatten()

for ax, col in zip(axes, columns_to_plot):
    ax.hist(df[col], color='red', bins=20)
    ax.set_xlabel(col)
    ax.set_ylabel('Frequency')
    ax.set_title(f'Histogram of {col}')
print(columns_to_plot[4])
# Hide any unused subplots
for ax in axes[len(columns_to_plot):]:
    ax.set_visible(False)

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Define the bins you want to use
bins = [4531.0, 10000, 50000, 100000, 500000, 1000000, 5000000, 10000000, 50000000, 100000000, 500000000, 1000000000, 3398480280]

barWidth = 0.3

r1 = np.arange(len(bins)-1)
r2 = [x + barWidth for x in r1]


# Use the cut function to bin the data for 1966
binned_data1 = pd.cut(df["1966"], bins, include_lowest=True)
# Use value_counts to get the frequency of each bin for 1966
frequency_table1 = binned_data1.value_counts().sort_index()

# Use the cut function to bin the data for 2022
binned_data2 = pd.cut(df["2022"], bins, include_lowest=True)
# Use value_counts to get the frequency of each bin for 2022
frequency_table2 = binned_data2.value_counts().sort_index()


# Plotting
plt.figure(figsize=(15, 6))

# Plot the bars for 1966
plt.bar(r1, frequency_table1, width=barWidth, color='skyblue', edgecolor='black', label='1966')

# Plot the bars for 2022 with a shift in x-position
plt.bar(r2 , frequency_table2, width=barWidth, color='orange', edgecolor='black', label='2022')
plt.xticks(r1, frequency_table1.index.astype(str), rotation=45)

# Adding titles and labels
plt.title('Histogram of Binned Data for 1966 and 2022')
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Manually extracted population data for Afghanistan
years = df.columns[4:]

# Calculate the total population for each year
total = df[years].sum()

# Calculate the differences between consecutive years
differences = [total[i] - total[i-1] for i in range(1, len(total))]
differences.insert(0, 0)

# Calculate the percentage change between consecutive years
percentage_change = [0] + [(differences[i] / total[i-1] * 100) if total[i-1] != 0 else 0 for i in range(1, len(total))]

# Create the index for the x-axis (years)
index = range(len(total))

# Create the horizontal bar chart
plt.figure(figsize=(7, 10))
bars = plt.barh(years, differences, color='skyblue')

# Adding titles and labels
plt.title('Yearly Population Change in World')
plt.xlabel('Population Change')
plt.ylabel('Year')

# Add percentage labels at the end of the bars
for bar, pc, diff in zip(bars, percentage_change, differences):
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height() / 2.0, f'{pc:.1f}%', ha='left', va='center')
    plt.text(diff, bar.get_y() + bar.get_height() / 2.0, f'{diff/1e6:.0f}M', ha='right', va='center', color='blue')
    
# Show the plot
plt.tight_layout()
plt.show()

# Calculate the average percentage change
average_percentage_change = sum(percentage_change) / len(percentage_change)

print(f"Average percentage change of population till now: {average_percentage_change:.2f}%")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming df is your DataFrame with data across years

# Create a dictionary to store country names and their percentage change
country_percentage_change = {}

# Calculate the change from 1960 to 2022
change = df['2022'] - df['1960']

# Calculate the percentage change
per_change = (change * 100) / df['1960']

# Store country names and their percentage change in the dictionary
for country, percentage in zip(df['Country Name'], per_change):
    country_percentage_change[country] = percentage

# Convert the dictionary to a pandas DataFrame
df_percentage_change = pd.DataFrame(list(country_percentage_change.items()), columns=['Country Name', 'Percentage Change'])

# Sort the DataFrame by percentage change in ascending order
df_percentage_change = df_percentage_change.sort_values(by='Percentage Change', ascending=True)

# Plot the horizontal bar chart
plt.figure(figsize=(11, 35))
bars = plt.barh(df_percentage_change['Country Name'], df_percentage_change['Percentage Change'], color='skyblue', edgecolor='black')

# Adding titles and labels
plt.title('Percentage Change in Population from 1960 to 2022 by Country')
plt.xlabel('Percentage Change')
plt.ylabel('Country')

# Add the percentage change text at the end of each bar
for bar in bars:
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height() / 2, f'{width:.2f}%', ha='left', va='center')

# Show the plot
plt.tight_layout()
plt.show()
```
